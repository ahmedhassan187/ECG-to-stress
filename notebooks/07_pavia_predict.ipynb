```python
import pandas as pd
import numpy as np

def load_pavia_hrv(path):
    df = pd.read_excel(path)
    return df


def build_dataset(df):
    """
    Convert wide HRV format into ML-ready dataset
    """

    feature_bases = [
        "HR", "SDNN", "rMSSD", "pNN50", "SE", "LF", "HF", "LFHF"
    ]

    X = []
    y = []

    for _, row in df.iterrows():

        for block in ["B1", "B2", "T1", "T2"]:

            sample = []

            for feat in feature_bases:
                col = f"{feat}_{block}"
                if col in df.columns:
                    sample.append(row[col])
                else:
                    sample.append(np.nan)

            # label: B = non-stress, T = stress
            label = 0 if "B" in block else 1

            X.append(sample)
            y.append(label)

    X = np.array(X, dtype=float)
    y = np.array(y, dtype=int)

    return X, y, feature_bases
```

```python
df = load_pavia_hrv("../data/HRV_Pavia_per_soggetto.xlsx")
X, y, feature_bases = build_dataset(df)
```

```python
def save_features_as_csv(X, y, feature_bases, path):
    """
    Save features and labels as CSV file
    """

    df = pd.DataFrame(X, columns=feature_bases)

    df.to_csv(path, index=False)
def save_labels_as_csv(y, path):
    """
    Save labels as CSV file
    """

    df = pd.DataFrame(y, columns=["label"])

    df.to_csv(path, index=False)
save_features_as_csv(X, y, feature_bases, "../data/pavia_features.csv")
save_labels_as_csv(y, "../data/pavia_labels.csv")
```
